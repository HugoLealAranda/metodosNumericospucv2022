```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la distribución de probabilidad
probabilidades = np.array([0.1, 0.2, 0.2, 0.3, 0.1, 0.1])
resultados = np.array([1, 2, 3, 4, 5, 6])

# Definimos la función de probabilidad condicional
def prob_condicional(p_x, p_y, p_x_y):
    return p_x_y / p_y

# Definimos la función de probabilidad conjunta
def prob_conjunta(p_x, p_y):
    return p_x * p_y

# Calculamos la probabilidad de obtener al menos una cara
p_al_menos_una_cara = 1 - probabilidades[0]

# Calculamos la probabilidad de obtener un resultado impar
p_impar = np.sum(probabilidades[0::2])

# Calculamos la probabilidad de obtener un resultado impar dado que se obtuvo al menos una cara
p_impar_al_menos_una_cara = prob_condicional(p_impar, p_al_menos_una_cara, prob_conjunta(p_impar, p_al_menos_una_cara))

# Mostramos el resultado
print("La probabilidad de obtener un resultado impar dado que se obtuvo al menos una cara es:", p_impar_al_menos_una_cara)

# Graficamos la distribución de probabilidad
plt.bar(resultados, probabilidades)
plt.title("Distribución de probabilidad")
plt.xlabel("Resultado")
plt.ylabel("Probabilidad")
plt.show()
```
