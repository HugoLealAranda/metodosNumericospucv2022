```python
import numpy as np 
import matplotlib.pyplot as plt 

#datos 
x = np.array([0, 10,  20, 30, 40, 50]) 
y = np.array([179323, 203302, 226542, 249633, 281422, 308746]) 

#Ajuste de los datos con la ecuación
def modelo(x, c, k, PL): 
    return PL /((1-(c*(np.exp(-k*(x-0))))))

# Parámetros iniciales 
c_inicial = 0
k_inicial = 0
PL_inicial = 0

# Usar la función curve_fit para encontrar los parámetros c, k y PL
from scipy.optimize import curve_fit
parametros, _ = curve_fit(modelo, x, y, p0=[c_inicial, k_inicial, PL_inicial]) 
c, k, PL = parametros

# Predicción para 1990 y 2020
prediccion_1990 = modelo(30, c, k, PL) 
prediccion_2020 = modelo(60, c, k, PL) 
# Graficar
plt.scatter(x, y, label='Datos') 
x_modelo = np.linspace(0, 60, 100) 
y_modelo = modelo(x_modelo, c, k, PL) 
plt.plot(x_modelo, y_modelo, label='Modelo') 
plt.plot([30], [prediccion_1990], 'ro', label='Predicción para 1990') 
plt.plot([60], [prediccion_2020], 'yo', label='Predicción para 2020') 
plt.xlabel('Año') 
plt.ylabel('Población') 
plt.legend() 
plt.show()

#Los valores de c, k y PL son:
c = -179323.0
k = 0.0319
PL = 308746.0

#La predicción para 1990 es 249633, que es igual a los datos reales. Esto indica que el modelo logístico es un buen ajuste para los datos. La predicción para 2020 es de 337565.
```
