```python
tiempo_inicial=time()
n = 1000
s = 0 
for i in range(n):
    for j in range(n):
        if i < j:
            s = s + i + j
tiempo_final=time()
tiempo_total=tiempo_final-tiempo_inicial
print("el tiempo que toma en ejecutar este codigo es de:", tiempo_total, "segundos", "y su eficiencia es de O(n**2)")
# el comando for in es un ciclo  que se considera un ciclo anidado en otro por eso N es de grado 2 
```

```python
tiempo_inicial=time()
n = 1000
s = 0
for i in range(n):
    s = s + i + sum(range(n))
tiempo_final=time()
tiempo_total=tiempo_final-tiempo_inicial
print("el tiempo que toma en ejecutar este codigo es de:", tiempo_total, "segundos", "y su eficiencia es de O(n**2)")
# el comando sum() es ciclo de suma, asi que se considera un ciclo anidado en otro por eso N es de grado 2 
```
